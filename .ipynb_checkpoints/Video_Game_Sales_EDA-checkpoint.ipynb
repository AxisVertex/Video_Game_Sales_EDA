<h1> Graphs and groupbys </h1>

```python
print(df.columns)
print(df.head())
```

```python
# Ditribution of games in platform
# plt.hist(df["Platform"])
# plt.xlabel("Platforms")
# plt.ylabel("Games")
# plt.title("Distribution of Games in Platforms")
# plt.show()

df["Platform"].value_counts().plot(kind="bar")
plt.xlabel("Platforms")
plt.ylabel("Games")
plt.title("Number of Games in each Platforms")
plt.show()
```

```python
df.groupby("Platform")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].sum().plot(kind="bar", figsize=(14, 7))
plt.title("Sales from each Platform in each Countries (in Millions)")
plt.xlabel("Platforms")
plt.ylabel("Sales (in Millions)")
plt.show()
```

```python
# amount of games in each genre
df["Genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Games")
plt.title("Number of Games in each Genre")
plt.show()
```

```python
# top 10 publishers
df["Publisher"].value_counts().head(10).plot(kind="bar")
plt.xlabel("Publishers")
plt.ylabel("Games")
plt.title("Amount of games published by the Top 10 Publishers")
plt.show()
```

```python
# amount of Games published in each year
df["Year"].value_counts().sort_index().plot(kind="bar", figsize=(12,6))
plt.xlabel("Year")
plt.ylabel("Games")
plt.title("Number of Games released per Year")
plt.show()
```

```python
# game genre over time
year_genre = df.groupby(["Year", "Genre"]).size().unstack(fill_value=0)

# year_genre.plot(kind="line", marker='*', figsize=(12, 8))
# plt.xlabel("Year")
# plt.ylabel("Amount")
# plt.title("Genres over Time")
# plt.show()


plt.figure(figsize=(12, 8))
for genre in year_genre:
    plt.plot(year_genre.index, year_genre[genre], label=genre, marker=".")

plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Top 5 Genre over Time")
plt.legend()

leg = plt.legend()
plt.show()

```

```python
# Total Sales (in millions)
df_subset = df[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].copy()
countries = ["NA", "EU", "JP", "Other", "Global"]
total_sales = {"Country":["NA", "EU", "JP", "Other", "Global"], "Total_Sales":[]}

for country in df_subset.columns:
    total_sales["Total_Sales"].append(df[country].sum())

total_sales = pd.DataFrame(total_sales)
total_sales

plt.bar(total_sales["Country"], total_sales["Total_Sales"])
plt.title("Total Sales Across Countries (in Millions)")
plt.xlabel("Countries")
plt.ylabel("Sales (in Millions)")
plt.show()
```

```python
df.groupby("Genre")["NA_Sales"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Sales per Genre in NA (in Millions)")
plt.xlabel("Genre")
plt.ylabel("Sale (in Millions)")
plt.show()
```

```python
df.groupby("Genre")["EU_Sales"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Sales per Genre in EU (in Millions)")
plt.xlabel("Genre")
plt.ylabel("Sale (in Millions)")
plt.show()
```

```python
df.groupby("Genre")["JP_Sales"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Sales per Genre in JP (in Millions)")
plt.xlabel("Genre")
plt.ylabel("Sale (in Millions)")
plt.show()
```

```python
df.groupby("Genre")["Global_Sales"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Sales per Genre Globally (in Millions)")
plt.xlabel("Genre")
plt.ylabel("Sale (in Millions)")
plt.show()
```

```python
df.groupby("Year")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].sum().plot(kind="line", figsize=(10,7), marker=".")
plt.title("Yearly Sales for each country (in Millions)")
plt.xlabel("Year")
plt.ylabel("Sales (in Millions")
plt.legend()
plt.show()
```
